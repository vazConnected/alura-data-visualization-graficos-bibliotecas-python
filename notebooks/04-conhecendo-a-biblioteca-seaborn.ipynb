```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Obter base de dados de imigração para o Canadá
df = pd.read_csv('../dados/canadian_immegration_data.csv')

# Transformar a coluna 'Country' em índice no data frame
df.set_index('Country', inplace=True)

# Obter os 10 países que mais imigraram para o Canadá
top_10 = df.sort_values(by='Total', ascending=False).head(10)

print("Top 10 países que mais imigraram para o Canadá:")
for index, pais in enumerate(top_10.index.tolist()):
    print(f"  {index+1}. {pais}")
```

```python
# Definir tema padrão para os gráficos
sns.set_theme(style='white')

# Criar gráfico de barras horizontal
ax = sns.barplot(data=top_10, y=top_10.index, x='Total',
                 orient='h', hue='Total', palette='dark:salmon', legend=False)
ax.grid(axis='x', linestyle='--', alpha=0.6)

# Modificar labels
ax.set_title('Top 10 países que mais imigraram para o Canadá\n(1980 a 2013)',
             loc='left', fontsize=14)
ax.set_xlabel('Total de imigrantes', fontsize=12)
ax.set_ylabel(ylabel='')

# Ajustar configurações do eixo x
ax.xaxis.set_major_locator(plt.MultipleLocator(50000))
ax.xaxis.set_tick_params(rotation=-90, labelsize=10, )
ax.xaxis.set_ticks_position('bottom')

# Remover bordas superior e direita do gráfico
sns.despine()

# Salvar imagem
figure = ax.get_figure()
figure.savefig('../resultados/top_10_paises_imigracao_canada.png',
               bbox_inches='tight', dpi=300)

plt.show()
```
